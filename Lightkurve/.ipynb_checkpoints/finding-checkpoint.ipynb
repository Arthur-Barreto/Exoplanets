```python
from lightkurve import search_targetpixelfile
import numpy as np
pixelFile = search_targetpixelfile('KIC 6922244', quarter=4).download()
# pegar a curva de luz da estrela, n√£o do planeta
```

```python
pixelFile
```

```python
lc = pixelFile.to_lightcurve(aperture_mask=pixelFile.pipeline_mask)
```

```python
lc.meta['MISSION']
```

```python
lc.meta['QUARTER']
```

```python
lc.time
```

```python
lc.plot();
```

```python
flat_lc = lc.flatten(window_length=401)
flat_lc.plot();
```

```python
folded_lc = flat_lc.fold(period=3.522)
folded_lc.plot();
```

```python
# rodando para todos os valores possiveis, com delta de 1d
lis_tempo = np.linspace(3.5,3.6,100)
for dt in lis_tempo:
    # 400d √© o tmax da kepler
    folded_lc = flat_lc.fold(period=dt)
    pg = lc.to_periodogram(oversample_factor=1)
    pg.plot(title=f'T = {dt} dias')
    
```

```python
binned_lc = folded_lc.bin(binsize=10)
binned_lc.plot();
```

```python
lc.remove_nans().flatten(window_length=401).fold(period=3.5225).bin(binsize=10).plot();
```

```python
import numpy as np
periodogram = flat_lc.to_periodogram(method="bls", period=np.arange(0.3, 10, 0.001))
periodogram.plot();
```

```python
best_fit_period = periodogram.period_at_max_power
print('Best fit period: {:.5f}'.format(best_fit_period))
```
