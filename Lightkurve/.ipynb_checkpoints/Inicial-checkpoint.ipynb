```python
from lightkurve import search_targetpixelfile
pixelfile = search_targetpixelfile("KIC 8462852", quarter=16).download();

# KIC 6922244 
# KIC 8462852
# KIC 10068024
```

```python
pixelfile.plot(frame=1);
```

```python
lc = pixelfile.to_lightcurve(aperture_mask='all');
```

```python
lc.time
```

```python
lc.flux
```

```python
lc.plot();
```

```python
pixelFile = search_targetpixelfile('KIC 6922244', quarter=4).download()
```

```python
lc = pixelFile.to_lightcurve(aperture_mask=pixelFile.pipeline_mask)
```

```python
lc.plot();
```

```python
flat_lc = lc.flatten(window_length=401)
flat_lc.plot();
```

```python
folded_lc = flat_lc.fold(period=3.5225)
folded_lc.plot();
```

```python
binned_lc = folded_lc.bin(binsize=10)
binned_lc.plot();
```

```python
lc.remove_nans().flatten(window_length=401).fold(period=3.5225).bin(binsize=10).plot();
```

```python
import numpy as np
periodogram = flat_lc.to_periodogram(method="bls", period=np.arange(-2, 100000, 0.001))
periodogram.plot();
```
